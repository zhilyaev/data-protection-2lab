#### Стиль отрисовки

```python
def background(x, key=key):
    df1 = pd.DataFrame('', index=x.index, columns=x.columns)
    style = 'background-color: lightgreen;'
    for row,col in key:
        df1.iloc[row, col] = style
    return df1
```

### Крутим ключи в этот раз, а не матрицы

```python
def index_at_0(coor=key, t=T[0]):
    s = []
    for row,col in coor:
        s.append((row,col))
    return s

def index_at_270(coor=key, t=T[0]):
    s = []
    for row,col in coor:
        s.append((9-col,row))
    return s

def index_at_180(coor=key, t=T[0]):
    s = []
    for row,col in coor:
        s.append((9-row, 9-col))
    return s

def index_at_90(coor=key, t=T[0]):
    s = []
    for row,col in coor:
        s.append((col, 9-row))
    return s
```

```python
T[0].style.apply(lambda x: background(x, index_at_0()), axis=None)
```

```python
key = [
    (0,1),(0,7),
    (1,2),(1,6),(1,7),(1,9),(1,10),
    (2,3),(2,11),
    (3,7),(3,8),(3,11),
    (4,10),
    (5,0),(5,2),(5,3),(5,8),(5,9),(5,10),
    (6,4),(6,6),
    (7,0),(7,2),(7,7),(7,9),
    (8,1),(8,7),(8,9),(8,11),
    (9,9),
    (10,3),
    (11,6),
]
len(src)/4 == len(key)
mask_key = [[False for x in range(12)] for y in range(12)]
for i in range(12):
    for j in range(12):
        if (i,j) in key:
            mask_key[i][j] = True
            
mask = np.array(mask_key)
```

```python
import random
ABC = "АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЯЭЮЯ"
ABC += ABC.lower()
ABC += '-,.'
TABLE = [[ABC[random.randint(0, len(ABC)-1)] for x in range(12)] for y in range(12)]
liner = 0
T = TABLE

for i,r in enumerate(T):
    for j,c in enumerate(r):
        if mask[i][j]:
            T[i][j] = src[liner]
            liner += 1

mask = np.rot90(mask, k=1)  
for i,r in enumerate(T):
    for j,c in enumerate(r):
        if mask[i][j]:
            T[i][j] = src[liner]
            liner += 1

T = pd.DataFrame(T)
T
```
